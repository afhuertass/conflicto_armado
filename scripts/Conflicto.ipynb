# Desaparecidos



```python
desaparecidos_tipo = []
for tipo , data in gp_victimas_des.groupby( level=0 ):
    #print(new_df.reset_index())
    #print( date )
    data = data.reset_index()
    
    sct = go.Scatter(
        x = data["ANNOH"] , 
        y = data["IdCaso"] , 
        name = tipo
    )
    desaparecidos_tipo.append( sct )
    
```

```python
layout = go.Layout(
    title = "Desaparecidos por año y situación actual de la victima" , 
    xaxis = dict( 
        title = "Año"
    ) , 
    yaxis = dict(
        title = "Desaparecidos"
    ) 
)
fig = go.Figure( data = desaparecidos_tipo , layout = layout )

iplot( fig, filename='desariciones_annoh') 
```

# Casos de masacres

Las masacres constituyeron una de las formas de violencia mas recurrentes en el pais durante los años del conflicto.

```python
df_casos_masacres = pd.read_csv("../data/casos_masacres.csv" , encoding = "utf-8")
```

```python
df_casos_masacres.head()
```

```python
df_casos_masacres["DESCRIPCION_PRESUNTO_RESPONSABLE"].tail()
```

```python
gp_masacresby_responsable = df_casos_masacres.groupby( by =  ["DESCRIPCION_PRESUNTO_RESPONSABLE"] )["NumeroVictimasCaso"].sum().reset_index()
```

```python
gp_masacresby_responsable.head()
```

```python
gp_masacresby_responsable["DESCRIPCION_PRESUNTO_RESPONSABLE"] = gp_masacresby_responsable["DESCRIPCION_PRESUNTO_RESPONSABLE"].apply(lambda x : "Autodefensas" if  "Autodefensas" in x  else x )
```

```python
gp_masacresby_responsable.head()
```

```python
gp_masacresby_responsable = gp_masacresby_responsable.groupby( by = [ "DESCRIPCION_PRESUNTO_RESPONSABLE"])["NumeroVictimasCaso"].sum()
```

```python
gp_masacresby_responsable = gp_masacresby_responsable.reset_index()

gp_masacresby_responsable = gp_masacresby_responsable.sort_values( by = "NumeroVictimasCaso" , ascending = False)
gp_masacresby_responsable.head()
```

```python
asesinatos_selectivos = go.Pie(
    labels=gp_masacresby_responsable["DESCRIPCION_PRESUNTO_RESPONSABLE"][:10] ,
    values= gp_masacresby_responsable["NumeroVictimasCaso"][:10] 
)
layout = go.Layout( 
    title = "Victimas de Masacres"
)

fig = go.Figure( data = [asesinatos_selectivos]  , layout = layout )

iplot( fig, filename='asesinatos_por_responsable') 
```

```python
df_casos_masacres_2 = df_casos_masacres[ df_casos_masacres["ANNOH"] != 0 ]
```

```python
gp_masacres_annoh_resp = df_casos_masacres_2.groupby( by =  ["DESCRIPCION_PRESUNTO_RESPONSABLE" , "ANNOH"  ]  )[ "NumeroVictimasCaso"].sum()
```

```python
gp_masacres_annoh_resp.head()
```

```python
masacres_resp = []
inx = 0
for tipo , data in gp_masacres_annoh_resp.groupby( level=0 ):
    #print(new_df.reset_index())
    #print( date )
    data = data.reset_index()
    
    sct = go.Scatter(
        x = data["ANNOH"] , 
        y = data["NumeroVictimasCaso"] , 
        name = tipo
    )
    inx = inx + 1 
    if inx > 10:
        break
    masacres_resp.append( sct )
```

```python
layout = go.Layout(
    title = "Victimas de masacres por año por presunto responsable" , 
    xaxis = dict( 
        title = "Año"
    ) , 
    yaxis = dict(
        title = "Desaparecidos"
    ) 
)
fig = go.Figure( data = masacres_resp , layout = layout )

iplot( fig, filename='masacres_resp_annoh') 
```
